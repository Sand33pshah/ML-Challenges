```python
dropping = ['Application_Receipt_Date', 'Applicant_BirthDate', 
            'Manager_DOJ','Manager_DoB','Manager_Num_Products2',
            'Manager_Business2','Manager_Current_Designation',
           'Manager_Joining_Designation']
train_data = train_data.drop(dropping, axis = 1)
train_data.head()
```

```python
test_data = test_data.drop(dropping, axis = 1)
test_data
```

```python
num_columns = ['Manager_Num_Application','Manager_Num_Coded',
               'Manager_Business','Manager_Num_Products']
column1 = ['Applicant_City_PIN','Manager_Grade','Manager_Status',
           'Manager_Gender','Applicant_Gender','Applicant_Marital_Status',
           'Applicant_Occupation','Applicant_Qualification']
```

```python
from sklearn.impute import SimpleImputer

imputer_num = SimpleImputer(missing_values = np.NaN, strategy = 'median')
train_data[num_columns] = imputer_num.fit_transform(train_data[num_columns])

```

```python

imputer1 = SimpleImputer(missing_values = np.NaN, strategy = 'most_frequent')
train_data[column1] = imputer1.fit_transform(train_data[column1])
```

```python
train_data.isnull().sum()
```

```python
imputer_num = SimpleImputer(missing_values = np.NaN, strategy = 'median')
test_data[num_columns] = imputer_num.fit_transform(test_data[num_columns])

imputer1 = SimpleImputer(missing_values = np.NaN, strategy = 'most_frequent')
test_data[column1] = imputer1.fit_transform(test_data[column1])
```

```python
test_data.isnull().sum()
```

```python
train_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
non_numerical = ['Applicant_Gender','Applicant_Marital_Status',
                  'Applicant_Occupation','Applicant_Qualification',
                 'Manager_Status','Manager_Gender']
for i in non_numerical:
    train_data[i] = le.fit_transform(train_data[i])
```

```python
for j in non_numerical:
    test_data[j] = le.fit_transform(test_data[j])
```

```python
train_data
```

```python
test_data
```

```python
from sklearn.preprocessing import StandardScaler

X = train_data.drop('Business_Sourced', axis = 1)
Y = train_data['Business_Sourced']


scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scaled_X, Y, train_size = 0.80,
                                                    shuffle = True, stratify = Y)
```

```python
y_test.head()
```

```python
y_train
```

```python
from sklearn.metrics import roc_auc_score, classification_report
```

```python
from sklearn.linear_model import LinearRegression as LiR

classifier5 = LiR(normalize = True)
classifier5.fit(x_train, y_train)
predicted_values = classifier5.predict(x_test)
print('ROC : ', roc_auc_score(y_test, predicted_values)*100, '%')
```

```python

scaler2 = StandardScaler()
scaled_test_data = scaler2.fit_transform(test_data)
```

```python
submission = pd.DataFrame()
submission['ID'] = test_id
submission['Business_Sourced'] = classifier5.predict(scaled_test_data)
submission.head(30)
```
